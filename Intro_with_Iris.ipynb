```python
from sklearn.datasets import load_iris
iris_data = load_iris()
```

```python
iris_data.keys()
```

```python
print(iris_data['DESCR']) #DESCR is the description of the dataset
```

```python
iris_data['target_names']
```

```python
iris_data['feature_names']
```

```python
type(iris_data['data'])
```

```python
iris_data['data'].shape
```

```python
X = iris_data['data']
y = iris_data['target']
```

```python
plt.scatter(X[:, 0], X[:, 2], c=y , s=60)
```

```python
#Assuming fetaures 0 and 2 give us the same information as the other features
```

```python
fig, ax = plt.subplots(3, 3, figsize=(15, 15))
plt.suptitle("iris pairplot")

for i in range(3):
  for j in range(3):
    ax[i, j].scatter(X[:, j], X[:, i+1], c=y , s=60)
    ax[i, j].set_xticks(())
    ax[i,j].set_yticks(())
    if i == 2:
      ax[i,j].set_xlabel(iris_data['feature_names'][j])
    if j == 0:
      ax[i,j].set_ylabel(iris_data['feature_names'][i])
    if j > i:
      ax[i,j].set_visible(False)
```
